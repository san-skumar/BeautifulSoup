```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
#url for scrapping data
url = "https://en.wikipedia.org/wiki/List_of_NBA_players_who_have_spent_their_entire_career_with_one_franchise"
```

```python
#Using get request to fetch the raw HTML content
html_content = requests.get(url).text
```

```python
#Parse the html content
soup = BeautifulSoup(html_content, 'lxml')

#print the parsed data of html
print(soup.prettify()) 
```

```python
#printing title text without the HTML tags 
soup.title.get_text()
```

```python
#printing all the links in the page along with its attributes
for link in soup.find_all('a'):
    print('Inner Text: {}'.format(link.text))
    print('Title: {}'.format(link.get('title')))
    print('href: {}\n'.format(link.get('href')))
```
